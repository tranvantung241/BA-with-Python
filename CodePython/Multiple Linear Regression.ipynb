```python
x=pd.read_csv('Colleges and Universities.csv')
x.head()
```

```python
import statsmodels.api as sm # import statsmodels 
## Y usually means our output/dependent variable
y = x["Graduation %"] 
## X usually means our input variables (or independent variables)
X = x[["Median SAT","Acceptance Rate","Expenditures/Student","Top 10% HS"] ]
X = sm.add_constant(X) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
predictions = model.predict(X)

# Print out the statistics
model.summary()
```
